```python
from urllib.parse import urlparse, parse_qs

def extract_video_id(video_url):
    parsed_url = urlparse(video_url)
    if "youtu.be" in parsed_url.netloc:
        return parsed_url.path[1:]
    elif "youtube.com" in parsed_url.netloc:
        query_params = parse_qs(parsed_url.query)
        return query_params.get('v', [''])[0]
    else:
        return None
```

```python
def transcribe_youtube_video(videoid):
    try:
        srt = YouTubeTranscriptApi.get_transcript(videoid)
        return srt
    except Exception as e:
        print(f"{str(e)}")
        return None
```

```python
transcripts = []
for video in videos:
    try:
        video_id = extract_video_id(video)
        print(video_id)
        transcript = transcribe_youtube_video(video_id)
        transcripts.append(transcript)
    except Exception as e:
        pass

len(transcripts)
```

```python
import csv
def save_transcripts_to_csv(transcripts, csv_filename):
    # Define the CSV file's header
    fieldnames = ['text', 'start', 'duration']

    # Open the CSV file for writing
    with open(csv_filename, mode='w', newline='') as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)


        writer.writeheader()

        for transcript in transcripts:
            if transcript is not None:
                for entry in transcript:
                    writer.writerow({'text': entry['text'], 'start': entry['start'], 'duration': entry['duration']})     
```

```python
filename= 'apple-vision-pro-reviews.csv'
save_transcripts_to_csv(transcripts, filename)
```

```python
import re

def case_folding_and_custom_cleaning(text):
    """
    Function to perform case folding and remove specific special characters (parentheses, square brackets, and some punctuations) 
    from the input text. It converts all characters in the text to lowercase.
    
    Parameters:
    text (str): The input text to be processed.
    
    Returns:
    str: The text converted to lowercase and cleaned of specified special characters.
    """
    # Convert text to lowercase
    text = text.lower()
    # Remove specific special characters using regular expressions
    text = re.sub(r'[()\[\]{}<>#@$!]', '', text)
    text = text.replace('speaker_00', '').strip()
    return text
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')

# Get NLTK English stop words
nltk_stopwords = set(stopwords.words('english'))
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load your transcript data into a pandas DataFrame
data = pd.read_csv('apple-vision-pro-reviews.csv') 
data['text'] = data['text'].apply(case_folding_and_custom_cleaning)

# Check the column names in your DataFrame
print(data.columns)

# Use the correct column name for the transcript data
text = ' '.join(data['text'])  # Replace 'correct_column_name' with the actual column name
words = word_tokenize(text.lower())  # Tokenize and convert to lowercase
```

```python
common_words = {',', '.', 'a', 'that', 'this', 'is', 'and', 'the', 'to', 'it', 'of','\'s','so','â™ª','in','on','i','you','like','but','for','can','with',
               'at','your','all','my','just','are','do','have','now','as','not',"'re", "n't","'m","'ve",
               }

common_words.update(nltk_stopwords)

len(common_words)
```

```python
filtered_words = [word for word in words if word not in common_words]

# Calculate word frequencies
word_freq = Counter(filtered_words)

# Most common words
most_common = word_freq.most_common(10)
print("Most common words:")
for word, freq in most_common:
    print(f"{word}: {freq}")

# Least common words
least_common = word_freq.most_common()[-10:]
print("\nLeast common words:")
for word, freq in least_common:
    print(f"{word}: {freq}")
```

```python
wordcloud = WordCloud(width=1200, height=720, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('AppleVision Pro Review')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a bar chart of word frequencies
plt.figure(figsize=(12, 6))
plt.bar(*zip(*most_common))  # Unzip most_common list
plt.xticks(rotation=45)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words')
plt.show()
```

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

bigram_measures = BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(words)
top_bigrams = finder.nbest(bigram_measures.raw_freq, 10)

print("Top 10 Bigrams:")
for bigram in top_bigrams:
    print(' '.join(bigram))
```

```python
from textblob import TextBlob

# Add a new column for sentiment analysis
data['Sentiment'] = data['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Add a new column 'Sentiment_Label' based on sentiment polarity
data['Sentiment_Label'] = data['Sentiment'].apply(lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral')

# Add a new column 'Sentiment_Code' encoding sentiment as 1 or 0
data['Sentiment_Code'] = data['Sentiment_Label'].apply(lambda x: 1 if x == 'Positive' else 0)

# Display the DataFrame with the new columns
print(data[['text', 'Sentiment_Label', 'Sentiment_Code']].head())
```

```python
columns_to_drop = ['start', 'duration']
data = data.drop(columns=columns_to_drop)
```
