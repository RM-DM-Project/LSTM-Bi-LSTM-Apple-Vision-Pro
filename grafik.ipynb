```python
import matplotlib.pyplot as plt
import numpy as np

# Data untuk metrik evaluasi
metrics = ['Loss(LSTM)', 'Loss(LSTM,K-Fold)', 'Loss(BiLSTM)', 'Loss(BiLSTM,K-Fold)']
scores = [0.04, 0.004, 0.04, 0.008]

# Warna untuk setiap bar
colors = ['blue', 'green', 'red', 'purple']

# Plotting bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(metrics, scores, color=colors)

# Menambahkan nilai di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    # Format angka berdasarkan kondisi: tampilkan dua desimal jika tidak ada trailing zero
    formatted_yval = f'{yval:.4f}' if yval < 0.01 else f'{yval:.2f}'
    plt.text(bar.get_x() + bar.get_width()/2, yval + max(scores) * 0.05, formatted_yval, ha='center', va='bottom')

# Judul dan label
plt.title('Evaluation Metrics')
plt.ylabel('Scores')

# Menambahkan batasan y-axis untuk lebih jelas
plt.ylim(0, max(scores) * 1.2)

# Tampilkan plot
plt.show()
```
